### 2.3 Removing duplicates

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
df = DataFrame({'column 1':[1,1,2,2,3,3,3],
               'column 2' : ['a', 'a', 'b', 'b', 'c', 'c', 'c'],
                'column 3' : ['A', 'A', 'B', 'B', 'C', 'C', 'C']
               })
df
```

```python
# use .duplicated() method to search each row of df to give True if a duplicate
df.duplicated() #Gives True on first duplicate
```

```python
# drop duplicates using .drop_duplicates()
df.drop_duplicates()
```

```python
# can also use this method to drop duplicates only on certain columns, as below
df = DataFrame({'column 1':[1,0,2,2,3,3,3],
               'column 2' : ['a', 'a', 'b', 'b', 'c', 'c', 'c'],
                'column 3' : ['A', 'A', 'B', 'B', 'C', 'D', 'C']
               })
df
```

```python
# pass in the column name like so
df.drop_duplicates('column 1')
```

```python
## Playing with my own examples:

# create a df
df = DataFrame(np.random.rand(25).reshape(5,5))
df
```

```python
# create some missing values
df.loc[:1, 3:4] = np.nan
df
```

```python
# create some duplicated values
df.loc[2:3, 1] = df.loc[2:3, 2]
df
```

```python
# remove columns with missing values
df.dropna().drop_duplicates(1)
```
